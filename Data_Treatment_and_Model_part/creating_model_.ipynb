```python
all_collided_satellites = pd.read_csv('collision_satellites.csv')
```

```python
all_sats_df = pd.concat([all_collided_satellites, all_non_colision_satellites])
```

```python
all_sats_df.info()
```

```python
columns_to_convert = ["perigee", "epoch", "mean_anomaly", "right_ascension", "mean_motion", "eccentricity", "bstar"]
all_sats_df[columns_to_convert] = all_sats_df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
all_sats_df["collided"] = all_sats_df["satellite_name"].isnull()
```

```python
pd.set_option('display.max_columns', None)
all_sats_df.head()
```

```python
for index, row in all_sats_df.iterrows():
    epoch_value = row['EPOCH']
    if isinstance(epoch_value, str):
        date_object = datetime.datetime.strptime(epoch_value, "%Y-%m-%dT%H:%M:%S.%f")
        timestamp = date_object.timestamp()
        all_sats_df.at[index, 'EPOCH'] = str(timestamp)
    elif isinstance(epoch_value, float):
        all_sats_df.at[index, 'EPOCH'] = str(epoch_value)
```

```python
all_sats_df['epoch'] = all_sats_df['epoch'].fillna(all_sats_df['EPOCH'])
all_sats_df['mean_anomaly'] = all_sats_df['mean_anomaly'].fillna(all_sats_df['MEAN_ANOMALY'])
all_sats_df['mean_motion'] = all_sats_df['mean_motion'].fillna(all_sats_df['MEAN_MOTION'])
all_sats_df['right_ascension'] = all_sats_df['right_ascension'].fillna(all_sats_df['RA_OF_ASC_NODE'])
all_sats_df['bstar'] = all_sats_df['bstar'].fillna(all_sats_df['BSTAR'])
all_sats_df['inclination'] = all_sats_df['inclination'].fillna(all_sats_df['INCLINATION'])
all_sats_df['eccentricity'] = all_sats_df['eccentricity'].fillna(all_sats_df['ECCENTRICITY'])
all_sats_df['satellite_name'] = all_sats_df['satellite_name'].fillna(all_sats_df['OBJECT_NAME'])
```

```python
all_sats_df.drop('MEAN_ANOMALY', inplace=True, axis=1)
all_sats_df.drop('MEAN_MOTION', inplace=True, axis=1)
all_sats_df.drop('RA_OF_ASC_NODE', inplace=True, axis=1)
all_sats_df.drop('BSTAR', inplace=True, axis=1)
all_sats_df.drop('INCLINATION', inplace=True, axis=1)
all_sats_df.drop('ECCENTRICITY', inplace=True, axis=1)
all_sats_df.drop('EPOCH', inplace=True, axis=1)
all_sats_df.drop('Unnamed: 0', inplace=True, axis=1)
all_sats_df.drop('OBJECT_NAME', inplace=True, axis=1)
```

```python
all_sats_df = all_sats_df.dropna(subset=['eccentricity'])
```

First model random forest regressor

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
x = all_sats_df[['inclination','mean_anomaly','right_ascension','mean_motion','eccentricity','bstar']]
y = all_sats_df['collided']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100, stratify=y)
```

```python
rf_params = {
   "n_estimators" : [50, 100, 200, 500],
   "max_depth" : [5, 10, 50],
   "random_state" : [1, 2, 5, 10]
}
```

```python
gs1_regressor = GridSearchCV(estimator=RandomForestRegressor(), param_grid=rf_params, scoring='accuracy', verbose=3)
gs1_regressor.fit(x_train, y_train)
```

```python
gs1_regressor.best_estimator_
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math
import matplotlib.pyplot as plt

y_predicted = gs1_regressor.best_estimator_.predict(x_test)
r2 = r2_score(y_test, y_predicted)
mae = mean_absolute_error(y_test, y_predicted)
mse = mean_squared_error(y_test, y_predicted)
rmse = math.sqrt(mse)

print("Mean Absolute Percentage Error (MAPE):", 1-mae)
print("Mean Squared Error (MSE):", mse)
print ("r2 score:", r2)
```

Model 2 classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_params = {
   "n_estimators" : [50, 100, 200, 500],
   "max_depth" : [5, 10, 50],
   "random_state" : [1, 2, 5, 10],
   "max_features": ["sqrt", "log2"]
}
```

```python
gs1_classifier = GridSearchCV(estimator=RandomForestClassifier(), param_grid=rf_params, scoring='accuracy', verbose=3)
gs1_classifier.fit(x_train, y_train)
```

```python
gs1_classifier.best_estimator_
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

y_predicted = gs1_classifier.best_estimator_.predict(x_test)
precision = accuracy_score(y_test, y_predicted)
f1 = f1_score(y_test, y_predicted)
roc_auc = roc_auc_score(y_test, y_predicted)

print("Precisin:", precision)
print("Puntaje F1:", f1)
print("rea bajo la curva ROC:", roc_auc)
```

```python
from sklearn.model_selection import cross_val_score

scores_train = cross_val_score(gs1_classifier, x_train, y_train, cv=5, scoring='accuracy')
scores_test = cross_val_score(gs1_classifier, x_test, y_test, cv=5, scoring='accuracy')


mean_score_train = scores_train.mean()
mean_score_test = scores_test.mean()

print("Puntuacin promedio en entrenamiento:", mean_score_train)
print("Puntuacin promedio en prueba:", mean_score_test)

if mean_score_train > mean_score_test:
    print("Hay indicios de overfitting")
else:
    print("No hay indicios de overfitting")
```
